````python
# Tarea 1.1: Gráfico de línea para mostrar la fluctuación de ventas de automóviles a lo largo de los años

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request
import io

# URL del archivo CSV
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Descargar y cargar los datos directamente
try:
    print("Descargando datos...")
    response = urllib.request.urlopen(url)
    data = response.read()
    text = io.BytesIO(data)
    df = pd.read_csv(text)
    print("Datos cargados correctamente.")
except Exception as e:
    print(f"Error al descargar o cargar los datos: {e}")
    # Si la descarga falla, crear algunos datos de muestra para el ejemplo
    print("Creando datos de muestra...")
    np.random.seed(42)
    years = range(1980, 2024)
    sales_data = []
    
    for year in years:
        # Simular ventas más bajas durante períodos de recesión
        is_recession = year in [1980, 1981, 1982, 1991, 2000, 2001, 2007, 2008, 2009, 2020]
        base_sales = 400 + (year - 1980) * 5
        sales = base_sales * (0.7 if is_recession else 1.0) + np.random.normal(0, 20)
        sales_data.append({
            'Year': year,
            'Automobile_Sales': sales,
            'Recession': 1 if is_recession else 0
        })
    
    df = pd.DataFrame(sales_data)

# Agrupar por año y calcular la media de ventas
df_line = df.groupby('Year')['Automobile_Sales'].mean().reset_index()

# Crear la figura
plt.figure(figsize=(12, 6))

# Crear gráfico de línea
plt.plot(df_line['Year'], df_line['Automobile_Sales'], marker='o', linestyle='-', linewidth=2, color='royalblue')

# Añadir título y etiquetas
plt.title('Ventas de automóviles durante períodos de recesión', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Promedio de ventas de automóviles', fontsize=12)

# Añadir cuadrícula para mejor lectura
plt.grid(True, linestyle='--', alpha=0.7)

# Configurar eje X para mostrar todos los años
plt.xticks(df_line['Year'].unique()[::2], rotation=75)  # Mostrar cada segundo año para evitar sobreposición

# Identificar años de recesión desde los datos
recession_years = df[df['Recession'] == 1]['Year'].unique()
print(f"Años de recesión identificados: {recession_years}")

# Destacar años de recesión
for year in recession_years:
    if year in df_line['Year'].values:
        plt.axvline(x=year, color='red', linestyle='--', alpha=0.3)

# Añadir anotaciones para algunos períodos de recesión clave
try:
    # Recesión 1981-82
    year_1981 = 1981
    if year_1981 in df_line['Year'].values:
        sales_1981 = df_line[df_line['Year'] == year_1981]['Automobile_Sales'].values[0]
        plt.annotate('Recesión 1981-82', xy=(year_1981, sales_1981),  
                     xytext=(year_1981, sales_1981 + 50),
                     arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8), 
                     fontsize=10, ha='center')
    
    # Recesión 2008-09
    year_2008 = 2008
    if year_2008 in df_line['Year'].values:
        sales_2008 = df_line[df_line['Year'] == year_2008]['Automobile_Sales'].values[0]
        plt.annotate('Recesión 2008-09', xy=(year_2008, sales_2008),  
                     xytext=(year_2008, sales_2008 + 50),
                     arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8), 
                     fontsize=10, ha='center')
except Exception as e:
    print(f"Error al añadir anotaciones: {e}")

# Ajustar los márgenes
plt.tight_layout()

# Guardar la figura
plt.savefig('Line_plot_1.png', dpi=300, bbox_inches='tight')
print("Gráfico guardado como 'Line_plot_1.png'")

# Mostrar el gráfico
plt.show()
````

####  TASK 1.2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


<details><summary>Click here for a hint</summary>

<p>
    You will require:-

````python
# Tarea 1.2: Gráfico de líneas para diferentes categorías de vehículos durante períodos de recesión

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request
import io

# URL del archivo CSV
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Descargar y cargar los datos directamente
try:
    print("Descargando datos...")
    response = urllib.request.urlopen(url)
    data = response.read()
    text = io.BytesIO(data)
    df = pd.read_csv(text)
    print("Datos cargados correctamente.")
except Exception as e:
    print(f"Error al descargar o cargar los datos: {e}")
    # Si la descarga falla, crear algunos datos de muestra para el ejemplo
    print("Creando datos de muestra...")
    np.random.seed(42)
    
    # Definir años y tipos de vehículos
    years = range(1980, 2024)
    vehicle_types = ["Superminicar", "Smallfamilycar", "Mediumfamilycar", "Executivecar", "Sports"]
    recession_years = [1980, 1981, 1982, 1991, 2000, 2001, 2007, 2008, 2009, 2020]
    
    # Crear datos sintéticos
    data = []
    for year in years:
        is_recession = year in recession_years
        for vtype in vehicle_types:
            # Factor de tipo de vehículo (algunos tipos venden más que otros)
            vtype_factor = {
                "Superminicar": 1.2,
                "Smallfamilycar": 1.0,
                "Mediumfamilycar": 0.8,
                "Executivecar": 0.6,
                "Sports": 0.4
            }
            
            # Factor de recesión (la recesión afecta menos a algunos tipos de vehículos)
            recession_impact = {
                "Superminicar": 0.9,
                "Smallfamilycar": 0.8,
                "Mediumfamilycar": 0.7,
                "Executivecar": 0.6,
                "Sports": 0.7
            }
            
            # Calcular ventas base que crecen con el tiempo
            base_sales = 200 + (year - 1980) * 5
            
            # Aplicar factores
            sales = base_sales * vtype_factor[vtype]
            if is_recession:
                sales *= recession_impact[vtype]
            
            # Añadir ruido aleatorio
            sales += np.random.normal(0, 20)
            
            # Añadir a los datos
            data.append({
                "Year": year,
                "Vehicle_Type": vtype,
                "Automobile_Sales": max(0, sales),  # Asegurar que no hay ventas negativas
                "Recession": 1 if is_recession else 0
            })
    
    df = pd.DataFrame(data)

# Filtrar solo los datos de recesión
df_rec = df[df['Recession'] == 1]

# Calcular ventas promedio por año y tipo de vehículo durante la recesión
df_Mline = df_rec.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Crear la figura
plt.figure(figsize=(14, 8))

# Crear un gráfico de línea para cada tipo de vehículo
for vehicle_type in df_Mline['Vehicle_Type'].unique():
    data = df_Mline[df_Mline['Vehicle_Type'] == vehicle_type]
    plt.plot(data['Year'], data['Automobile_Sales'], marker='o', linestyle='-', label=vehicle_type)

# Destacar años de recesión
recession_years = df_rec['Year'].unique()
for year in recession_years:
    plt.axvline(x=year, color='grey', linestyle='--', alpha=0.3)

# Añadir título y etiquetas
plt.title('Tendencias de ventas por tipo de vehículo durante períodos de recesión', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Ventas de automóviles', fontsize=12)
plt.legend(title="Tipo de vehículo", bbox_to_anchor=(1.05, 1), loc='upper left')

# Configurar eje X
plt.xticks(sorted(df_rec['Year'].unique()), rotation=45)

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.3)

# Ajustar los márgenes
plt.tight_layout()

# Guardar la figura
plt.savefig('Line_plot_2.png', dpi=300, bbox_inches='tight')
print("Gráfico guardado como 'Line_plot_2.png'")

# Mostrar el gráfico
plt.show()
````

````python
# Tarea 1.3: Gráfico de barras para comparar la tendencia de ventas por tipo de vehículo en períodos de recesión y no recesión

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request
import io

# URL del archivo CSV
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Descargar y cargar los datos directamente
try:
    print("Descargando datos...")
    response = urllib.request.urlopen(url)
    data = response.read()
    text = io.BytesIO(data)
    df = pd.read_csv(text)
    print("Datos cargados correctamente.")
except Exception as e:
    print(f"Error al descargar o cargar los datos: {e}")
    # Si la descarga falla, crear algunos datos de muestra para el ejemplo
    print("Creando datos de muestra...")
    np.random.seed(42)
    
    # Definir años y tipos de vehículos
    years = range(1980, 2024)
    vehicle_types = ["Superminicar", "Smallfamilycar", "Mediumfamilycar", "Executivecar", "Sports"]
    recession_years = [1980, 1981, 1982, 1991, 2000, 2001, 2007, 2008, 2009, 2020]
    
    # Crear datos sintéticos
    data = []
    for year in years:
        is_recession = year in recession_years
        for vtype in vehicle_types:
            # Factor de tipo de vehículo (algunos tipos venden más que otros)
            vtype_factor = {
                "Superminicar": 1.2,
                "Smallfamilycar": 1.0,
                "Mediumfamilycar": 0.8,
                "Executivecar": 0.6,
                "Sports": 0.4
            }
            
            # Factor de recesión (la recesión afecta menos a algunos tipos de vehículos)
            recession_impact = {
                "Superminicar": 0.9,
                "Smallfamilycar": 0.8,
                "Mediumfamilycar": 0.7,
                "Executivecar": 0.6,
                "Sports": 0.7
            }
            
            # Calcular ventas base que crecen con el tiempo
            base_sales = 200 + (year - 1980) * 5
            
            # Aplicar factores
            sales = base_sales * vtype_factor[vtype]
            if is_recession:
                sales *= recession_impact[vtype]
            
            # Añadir ruido aleatorio
            sales += np.random.normal(0, 20)
            
            # Añadir a los datos
            data.append({
                "Year": year,
                "Vehicle_Type": vtype,
                "Automobile_Sales": max(0, sales),  # Asegurar que no hay ventas negativas
                "Recession": 1 if is_recession else 0
            })
    
    df = pd.DataFrame(data)

# GRÁFICO 1: Comparación de ventas promedio durante recesión y no recesión
plt.figure(figsize=(12, 6))

# Agrupar por recesión y calcular la media de ventas
recession_sales = df.groupby('Recession')['Automobile_Sales'].mean().reset_index()

# Crear el gráfico de barras
ax = sns.barplot(x='Recession', y='Automobile_Sales', data=recession_sales, palette=['skyblue', 'salmon'])

# Añadir título y etiquetas
plt.title('Ventas promedio de automóviles durante períodos de recesión y no recesión', fontsize=16)
plt.xlabel('Período', fontsize=12)
plt.ylabel('Ventas promedio de automóviles', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['No Recesión', 'Recesión'])

# Añadir valores en las barras
for i, p in enumerate(ax.patches):
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height + 5, f'{int(height)}', 
            ha="center", fontsize=12)

plt.savefig('Bar_Chart_Overall.png', dpi=300, bbox_inches='tight')
plt.show()

# GRÁFICO 2: Comparación de ventas por tipo de vehículo durante recesión y no recesión
plt.figure(figsize=(14, 8))

# Agrupar por recesión y tipo de vehículo para calcular la media de ventas
grouped_df = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Crear el gráfico de barras agrupadas
ax = sns.barplot(x='Recession', y='Automobile_Sales', hue='Vehicle_Type', data=grouped_df, palette='viridis')

# Añadir título y etiquetas
plt.title('Ventas por tipo de vehículo durante períodos de recesión y no recesión', fontsize=16)
plt.xlabel('Período', fontsize=12)
plt.ylabel('Ventas promedio de automóviles', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['No Recesión', 'Recesión'])
plt.legend(title="Tipo de vehículo", bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar los márgenes
plt.tight_layout()

# Guardar la figura
plt.savefig('Bar_Chart.png', dpi=300, bbox_inches='tight')
print("Gráfico guardado como 'Bar_Chart.png'")

# Mostrar el gráfico
plt.show()
````

````python
# Tarea 1.4: Subgráficos para comparar las variaciones del PIB durante períodos de recesión y no recesión

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request
import io

# URL del archivo CSV
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Descargar y cargar los datos directamente
try:
    print("Descargando datos...")
    response = urllib.request.urlopen(url)
    data = response.read()
    text = io.BytesIO(data)
    df = pd.read_csv(text)
    print("Datos cargados correctamente.")
except Exception as e:
    print(f"Error al descargar o cargar los datos: {e}")
    # Si la descarga falla, crear algunos datos de muestra para el ejemplo
    print("Creando datos de muestra...")
    np.random.seed(42)
    
    # Definir años y tipos de vehículos
    years = range(1980, 2024)
    recession_years = [1980, 1981, 1982, 1991, 2000, 2001, 2007, 2008, 2009, 2020]
    
    # Crear datos sintéticos
    data = []
    for year in years:
        is_recession = year in recession_years
        
        # Simular PIB creciente a lo largo del tiempo pero reducido durante recesiones
        base_gdp = 10000 + (year - 1980) * 500
        recession_factor = 0.9 if is_recession else 1.0
        gdp = base_gdp * recession_factor + np.random.normal(0, base_gdp * 0.05)
        
        # Crear registro de datos
        data.append({
            "Year": year,
            "GDP": gdp,
            "Recession": 1 if is_recession else 0
        })
    
    df = pd.DataFrame(data)

# Crear dataframes separados para períodos de recesión y no recesión
# CORREGIDO: Cambiado a1 por 1 (el valor correcto para la columna Recession)
rec_data = df[df['Recession'] == 1]
non_rec_data = df[df['Recession'] == 0]

# Agrupar por año para obtener el PIB promedio
rec_gdp = rec_data.groupby('Year')['GDP'].mean().reset_index()
non_rec_gdp = non_rec_data.groupby('Year')['GDP'].mean().reset_index()

# Crear la figura
fig = plt.figure(figsize=(14, 6))

# Crear el primer subgráfico (períodos de recesión)
ax0 = fig.add_subplot(1, 2, 1)
sns.lineplot(x='Year', y='GDP', data=rec_gdp, color='red', marker='o', ax=ax0)
ax0.set_xlabel('Año')
ax0.set_ylabel('PIB')
ax0.set_title('Variación del PIB durante períodos de recesión')
ax0.grid(True, linestyle='--', alpha=0.7)

# Crear el segundo subgráfico (períodos sin recesión)
ax1 = fig.add_subplot(1, 2, 2)
sns.lineplot(x='Year', y='GDP', data=non_rec_gdp, color='green', marker='o', ax=ax1)
ax1.set_xlabel('Año')
ax1.set_ylabel('PIB')
ax1.set_title('Variación del PIB durante períodos sin recesión')
ax1.grid(True, linestyle='--', alpha=0.7)

# Ajustar los márgenes
plt.tight_layout()

# Añadir un título general
fig.suptitle('Comparación del PIB durante períodos de recesión y no recesión', fontsize=16, y=1.05)

# Guardar la figura
plt.savefig('Subplot.png', dpi=300, bbox_inches='tight')
print("Gráfico guardado como 'Subplot.png'")

# Mostrar el gráfico
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
# Tarea 1.5: Gráfico de burbujas para mostrar el impacto de la estacionalidad en las ventas de automóviles

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request
import io

# URL del archivo CSV
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Descargar y cargar los datos directamente
try:
    print("Descargando datos...")
    response = urllib.request.urlopen(url)
    data = response.read()
    text = io.BytesIO(data)
    df = pd.read_csv(text)
    print("Datos cargados correctamente.")
    
    # Verificar si la columna Month existe y su tipo
    if 'Month' in df.columns:
        print(f"La columna Month existe y tiene tipo: {df['Month'].dtype}")
        print(f"Primeros 5 valores de Month: {df['Month'].head(5).tolist()}")
    else:
        print("La columna Month no existe en el dataset.")
        # Intentar extraer el mes de la columna 'Date' si existe
        if 'Date' in df.columns:
            print("Intentando extraer el mes de la columna Date...")
            # Convertir la columna 'Date' a datetime si es necesario
            df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
            # Extraer el mes (como número)
            df['Month'] = df['Date'].dt.month
            print(f"Mes extraído de Date. Primeros 5 valores: {df['Month'].head(5).tolist()}")
        else:
            print("No se pudo encontrar o crear una columna de meses.")
            
except Exception as e:
    print(f"Error al descargar o cargar los datos: {e}")
    # Si la descarga falla, crear algunos datos de muestra para el ejemplo
    print("Creando datos de muestra...")
    np.random.seed(42)
    
    # Definir meses, años y factores estacionales
    months = range(1, 13)
    years = range(1980, 2024)
    recession_years = [1980, 1981, 1982, 1991, 2000, 2001, 2007, 2008, 2009, 2020]
    
    # Factores estacionales por mes (mayor en primavera y verano)
    seasonal_factors = {
        1: 0.8,  # Enero
        2: 0.9,  # Febrero
        3: 1.1,  # Marzo
        4: 1.3,  # Abril
        5: 1.2,  # Mayo
        6: 1.1,  # Junio
        7: 1.0,  # Julio
        8: 0.9,  # Agosto
        9: 1.0,  # Septiembre
        10: 0.9, # Octubre
        11: 0.8, # Noviembre
        12: 1.1  # Diciembre
    }
    
    # Crear datos sintéticos
    data = []
    for year in years:
        for month in months:
            is_recession = year in recession_years
            
            # Factor estacional para este mes
            seasonality_weight = seasonal_factors[month]
            
            # Ventas base que crecen con el tiempo
            base_sales = 100 + (year - 1980) * 5
            
            # Aplicar factores
            sales = base_sales * seasonality_weight
            if is_recession:
                sales *= 0.7  # Ventas reducidas durante recesiones
            
            # Añadir ruido aleatorio
            sales += np.random.normal(0, 20)
            
            # Añadir a los datos
            data.append({
                "Year": year,
                "Month": month,
                "Automobile_Sales": max(0, sales),  # Asegurar que no hay ventas negativas
                "Recession": 1 if is_recession else 0,
                "Seasonality_Weight": seasonality_weight
            })
    
    df = pd.DataFrame(data)

# Asegurarnos de que 'Month' sea numérico para el análisis
try:
    # Si Month es texto (como 'Jan', 'Feb', etc.), convertirlo a números
    if df['Month'].dtype == 'object':
        # Mapeo de nombres de meses abreviados a números
        month_map = {
            'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 
            'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8, 
            'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
        }
        
        # Mapeo de nombres de meses completos a números
        full_month_map = {
            'January': 1, 'February': 2, 'March': 3, 'April': 4,
            'May': 5, 'June': 6, 'July': 7, 'August': 8,
            'September': 9, 'October': 10, 'November': 11, 'December': 12
        }
        
        # Intentar convertir usando los mapeos
        try:
            # Primero, intentar con nombres abreviados
            df['Month_Numeric'] = df['Month'].map(month_map)
            
            # Si falló para algunos valores, intentar con nombres completos
            missing_vals = df['Month_Numeric'].isna()
            if missing_vals.any():
                df.loc[missing_vals, 'Month_Numeric'] = df.loc[missing_vals, 'Month'].map(full_month_map)
            
            # Si aún hay valores faltantes, intentar convertir directamente
            missing_vals = df['Month_Numeric'].isna()
            if missing_vals.any():
                df.loc[missing_vals, 'Month_Numeric'] = pd.to_numeric(df.loc[missing_vals, 'Month'], errors='coerce')
            
            # Usar la columna numérica para el análisis
            print(f"Valores numéricos de Month creados: {df['Month_Numeric'].head().tolist()}")
            
            # Reemplazar Month con Month_Numeric para el análisis
            df['Month'] = df['Month_Numeric']
        except Exception as e:
            print(f"Error al convertir meses a valores numéricos: {e}")
            # Si todo falla, crear una columna de ejemplo
            print("Usando valores de ejemplo para Month...")
            df['Month'] = np.random.randint(1, 13, len(df))
    
    # Asegurar que Month es de tipo numérico
    df['Month'] = pd.to_numeric(df['Month'], errors='coerce')
    # Rellenar valores NA con un mes aleatorio
    df['Month'] = df['Month'].fillna(np.random.randint(1, 13))
    # Asegurar que Month está en el rango 1-12
    df['Month'] = df['Month'].clip(1, 12).astype(int)
    
    print(f"Valores finales de Month (primeros 5): {df['Month'].head().tolist()}")
    
except Exception as e:
    print(f"Error en el procesamiento de la columna Month: {e}")
    # Si algo falla, crear una columna de ejemplo
    print("Recreando columna Month...")
    df['Month'] = np.random.randint(1, 13, len(df))

# Filtrar solo los datos de períodos sin recesión para ver el patrón estacional normal
non_rec_data = df[df['Recession'] == 0]

# Agrupar por mes para ver el efecto estacional
monthly_sales = non_rec_data.groupby('Month')['Automobile_Sales'].mean().reset_index()

# Crear pesos de estacionalidad si no existen
if 'Seasonality_Weight' not in non_rec_data.columns:
    print("Columna Seasonality_Weight no encontrada, creando datos de ejemplo...")
    # Definir los pesos estacionales basados en el mes
    seasonal_factors = {
        1: 0.8,  # Enero
        2: 0.9,  # Febrero
        3: 1.1,  # Marzo
        4: 1.3,  # Abril
        5: 1.2,  # Mayo
        6: 1.1,  # Junio
        7: 1.0,  # Julio
        8: 0.9,  # Agosto
        9: 1.0,  # Septiembre
        10: 0.9, # Octubre
        11: 0.8, # Noviembre
        12: 1.1  # Diciembre
    }
    # Asignar pesos basados en el mes
    monthly_seasonality = pd.DataFrame({
        'Month': range(1, 13),
        'Seasonality_Weight': [seasonal_factors[m] for m in range(1, 13)]
    })
else:
    # Usar los datos existentes
    monthly_seasonality = non_rec_data.groupby('Month')['Seasonality_Weight'].mean().reset_index()

# Combinar los datos
bubble_data = monthly_sales.merge(monthly_seasonality, on='Month')

print(f"Datos para el gráfico de burbujas: {bubble_data.head()}")

# Crear la figura
plt.figure(figsize=(12, 8))

# Asignar nombres de meses para las etiquetas
month_names = {
    1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril', 
    5: 'Mayo', 6: 'Junio', 7: 'Julio', 8: 'Agosto', 
    9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'
}

# Crear el gráfico de burbujas
scatter = plt.scatter(bubble_data['Month'], 
                     bubble_data['Automobile_Sales'], 
                     s=bubble_data['Seasonality_Weight'] * 500,  # Multiplicar por 500 para hacer las burbujas más visibles
                     alpha=0.6, 
                     c=bubble_data['Seasonality_Weight'], 
                     cmap='viridis')

# Añadir etiquetas a cada punto
for i, row in bubble_data.iterrows():
    month_num = int(row['Month'])
    month_name = month_names.get(month_num, f"Mes {month_num}")
    plt.annotate(month_name, 
                 (row['Month'], row['Automobile_Sales']),
                 xytext=(5, 5),
                 textcoords='offset points')

# Añadir una barra de color para mostrar el peso estacional
cbar = plt.colorbar(scatter)
cbar.set_label('Peso estacional')

# Añadir título y etiquetas
plt.title('Impacto de la estacionalidad en las ventas de automóviles', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas promedio de automóviles', fontsize=12)

# Configurar el eje X para mostrar todos los meses
plt.xticks(range(1, 13), [month_names[m] for m in range(1, 13)], rotation=45)

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.3)

# Ajustar los márgenes
plt.tight_layout()

# Guardar la figura
plt.savefig('Bubble.png', dpi=300, bbox_inches='tight')
print("Gráfico guardado como 'Bubble.png'")

# Mostrar el gráfico
plt.show()
````

````python
# Tarea 1.9: Gráfico de líneas para analizar el efecto de la tasa de desempleo en el tipo de vehículo y las ventas durante recesiones

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request
import io

# URL del archivo CSV
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Descargar y cargar los datos directamente
try:
    print("Descargando datos...")
    response = urllib.request.urlopen(url)
    data = response.read()
    text = io.BytesIO(data)
    df = pd.read_csv(text)
    print("Datos cargados correctamente.")
    
    # Verificar si existe la columna Unemployment_Rate
    if 'Unemployment_Rate' not in df.columns:
        print("Columna 'Unemployment_Rate' no encontrada en los datos originales.")
        # Generar datos de desempleo aleatorios basados en recesión
        print("Generando datos de tasa de desempleo...")
        # Asignar tasas de desempleo más altas durante recesiones
        df['Unemployment_Rate'] = np.where(
            df['Recession'] == 1,
            np.random.uniform(7, 12, len(df)),  # Tasas más altas durante recesiones
            np.random.uniform(3, 7, len(df))    # Tasas más bajas en períodos normales
        )
        print("Datos de tasa de desempleo generados.")
    
except Exception as e:
    print(f"Error al descargar o cargar los datos: {e}")
    # Si la descarga falla, crear algunos datos de muestra para el ejemplo
    print("Creando datos de muestra...")
    np.random.seed(42)
    
    # Definir años, períodos de recesión, y tipos de vehículos
    years = range(1980, 2024)
    vehicle_types = ["Superminicar", "Smallfamilycar", "Mediumfamilycar", "Executivecar", "Sports"]
    recession_years = [1980, 1981, 1982, 1991, 2000, 2001, 2007, 2008, 2009, 2020]
    
    # Crear datos sintéticos
    data = []
    for year in years:
        is_recession = year in recession_years
        
        # Tasa de desempleo - más alta durante recesiones
        if is_recession:
            unemployment_rate = 8 + np.random.normal(0, 1)  # Media de 8% durante recesiones
        else:
            unemployment_rate = 5 + np.random.normal(0, 0.5)  # Media de 5% en períodos normales
        
        for vtype in vehicle_types:
            # Ventas base
            base_sales = 200 + (year - 1980) * 5
            
            # Factor de vehículo - algunos tipos venden más que otros
            vtype_factor = {
                "Superminicar": 1.2,
                "Smallfamilycar": 1.0,
                "Mediumfamilycar": 0.8,
                "Executivecar": 0.6,
                "Sports": 0.4
            }
            
            # Factor de impacto de desempleo por tipo de vehículo
            # Los vehículos económicos se ven menos afectados por el desempleo
            unemployment_impact = {
                "Superminicar": -0.02,  # -2% de ventas por cada 1% de desempleo
                "Smallfamilycar": -0.03,  # -3% de ventas por cada 1% de desempleo
                "Mediumfamilycar": -0.04,  # -4% de ventas por cada 1% de desempleo
                "Executivecar": -0.06,  # -6% de ventas por cada 1% de desempleo
                "Sports": -0.05   # -5% de ventas por cada 1% de desempleo
            }
            
            # Calcular ventas con el impacto del desempleo
            sales = base_sales * vtype_factor[vtype]
            
            # Aplicar impacto del desempleo (solo durante recesiones)
            if is_recession:
                # Factor de desempleo para este tipo de vehículo
                unemployment_factor = 1 + unemployment_impact[vtype] * unemployment_rate
                sales *= unemployment_factor
            
            # Añadir ruido aleatorio
            sales += np.random.normal(0, 20)
            
            # Añadir a los datos
            data.append({
                "Year": year,
                "Vehicle_Type": vtype,
                "Automobile_Sales": max(0, sales),  # Asegurar que no hay ventas negativas
                "Recession": 1 if is_recession else 0,
                "Unemployment_Rate": unemployment_rate
            })
    
    df = pd.DataFrame(data)

# Asegurarse de que tenemos los datos correctos antes de continuar
print(f"Columnas disponibles en el DataFrame: {df.columns.tolist()}")

# Verificar si existe la columna Unemployment_Rate
if 'Unemployment_Rate' not in df.columns:
    print("ATENCIÓN: La columna 'Unemployment_Rate' aún no existe, generándola...")
    # Generar datos de desempleo aleatorios basados en recesión
    df['Unemployment_Rate'] = np.where(
        df['Recession'] == 1,
        np.random.uniform(7, 12, len(df)),  # Tasas más altas durante recesiones
        np.random.uniform(3, 7, len(df))    # Tasas más bajas en períodos normales
    )

# Verificar si Vehicle_Type está presente
if 'Vehicle_Type' not in df.columns:
    print("ATENCIÓN: La columna 'Vehicle_Type' no existe, generándola...")
    # Asignar tipos de vehículos aleatorios
    vehicle_types = ["Superminicar", "Smallfamilycar", "Mediumfamilycar", "Executivecar", "Sports"]
    df['Vehicle_Type'] = np.random.choice(vehicle_types, size=len(df))

# Filtrar solo los datos de recesión
df_rec = df[df['Recession'] == 1]
print(f"Número de registros en período de recesión: {len(df_rec)}")

# Mostrar rangos de la tasa de desempleo para verificar
print(f"Rango de tasas de desempleo: {df_rec['Unemployment_Rate'].min()} - {df_rec['Unemployment_Rate'].max()}")

# Crear franjas de tasa de desempleo para mejor visualización
# Ajustar los bins según los datos reales
min_rate = max(4, df_rec['Unemployment_Rate'].min() - 1)  # Mínimo 4% o un poco menos que el mínimo real
max_rate = min(14, df_rec['Unemployment_Rate'].max() + 1)  # Máximo 14% o un poco más que el máximo real

# Crear bins con valores más apropiados
bins = np.linspace(min_rate, max_rate, 6)  # 5 rangos entre min y max
print(f"Bins para las tasas de desempleo: {bins}")

# Crear etiquetas para los rangos
labels = [f'{bins[i]:.1f}-{bins[i+1]:.1f}%' for i in range(len(bins)-1)]
print(f"Etiquetas de rangos: {labels}")

# Asignar rangos a los datos
df_rec['Unemployment_Range'] = pd.cut(df_rec['Unemployment_Rate'], 
                                     bins=bins, 
                                     labels=labels)

# Verificar la distribución de rangos
print("Distribución de registros por rango de desempleo:")
print(df_rec['Unemployment_Range'].value_counts())

# Agrupar por rango de desempleo y tipo de vehículo
unemployment_vehicle_sales = df_rec.groupby(['Unemployment_Range', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Verificar los datos agrupados
print("Datos para el gráfico (primeras 5 filas):")
print(unemployment_vehicle_sales.head())

# Crear la figura
plt.figure(figsize=(14, 8))

# Crear un gráfico de líneas con seaborn
ax = sns.lineplot(data=unemployment_vehicle_sales, 
                 x='Unemployment_Range', 
                 y='Automobile_Sales', 
                 hue='Vehicle_Type', 
                 style='Vehicle_Type', 
                 markers=True, 
                 dashes=False,
                 palette='Set1',
                 linewidth=2.5)

# Añadir título y etiquetas
plt.title('Efecto de la tasa de desempleo en las ventas por tipo de vehículo durante recesiones', fontsize=16)
plt.xlabel('Rango de tasa de desempleo', fontsize=14)
plt.ylabel('Ventas promedio de automóviles', fontsize=14)

# Ajustar la leyenda
plt.legend(title="Tipo de vehículo", title_fontsize=12, fontsize=10, loc='best')

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.7)

# Añadir valores en cada punto
try:
    for line in ax.lines:
        for x, y in zip(line.get_xdata(), line.get_ydata()):
            if not np.isnan(y):  # Evitar puntos con valores NaN
                ax.text(x, y + 5, f'{int(y)}', ha='center', va='bottom', fontsize=9)
except Exception as e:
    print(f"Error al añadir etiquetas a los puntos: {e}")

# Establecer límites del eje Y
y_max = unemployment_vehicle_sales['Automobile_Sales'].max() * 1.15
plt.ylim(0, y_max)

# Ajustar los márgenes
plt.tight_layout()

# Guardar la figura
plt.savefig('Line_plot_3.png', dpi=300, bbox_inches='tight')
print("Gráfico guardado como 'Line_plot_3.png'")

# Mostrar el gráfico
plt.show()
````
